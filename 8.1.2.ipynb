```python
from pyomo.environ import*
from pyomo.opt import*

model=ConcreteModel()

model.x_11=Var(within=NonNegativeReals)
model.x_12=Var(within=NonNegativeReals)
model.x_13=Var(within=NonNegativeReals)
model.x_14=Var(within=NonNegativeReals)
model.x_21=Var(within=NonNegativeReals)
model.x_22=Var(within=NonNegativeReals)
model.x_23=Var(within=NonNegativeReals)
model.x_24=Var(within=NonNegativeReals)
model.x_31=Var(within=NonNegativeReals)
model.x_32=Var(within=NonNegativeReals)
model.x_33=Var(within=NonNegativeReals)
model.x_34=Var(within=NonNegativeReals)


model.obj= Objective(expr= model.x_11*(100+0.5*800)+model.x_12*(100+0.5*1300)+model.x_13*(100+0.5*400)+
                     model.x_14*(100+0.5*700)+model.x_21*(100+0.5*1100)+model.x_22*(100+0.5*1400)+
                     model.x_23*(100+0.5*600)+model.x_24*(100+0.5*1000)+model.x_31*(100+0.5*600)+
                     model.x_32*(100+0.5*1200)+model.x_33*(100+0.5*800)+model.x_34*(100+0.5*900)  , 
                     sense=minimize)
 
model.con1= Constraint(expr= model.x_11 + model.x_12 + model.x_13 ==12)
model.con2= Constraint(expr= model.x_21 + model.x_22 + model.x_23 ==17)
model.con3= Constraint(expr= model.x_31 + model.x_32 + model.x_33 ==11) 
model.con4= Constraint(expr= model.x_11 + model.x_21 + model.x_31 ==10)
model.con5= Constraint(expr= model.x_12 + model.x_22 + model.x_23 ==10)
model.con6= Constraint(expr= model.x_13 + model.x_23 + model.x_33 ==10)
model.con7= Constraint(expr= model.x_14 + model.x_24 + model.x_34 ==10)

Solver= SolverFactory('glpk')
results= Solver.solve(model)
print(results)

print('value of x11 is', value(model.x_11))
print('value of x12 is', value(model.x_12))
print('value of x13 is', value(model.x_13))
print('value of x14 is', value(model.x_14))
print('value of x21 is', value(model.x_21))
print('value of x22 is', value(model.x_22))
print('value of x23 is', value(model.x_23))
print('value of x24 is', value(model.x_24))
print('value of x31 is', value(model.x_31))
print('value of x32 is', value(model.x_32))
print('value of x33 is', value(model.x_33))
print('value of x34 is', value(model.x_34))
print('min value is', value(model.obj))
```
