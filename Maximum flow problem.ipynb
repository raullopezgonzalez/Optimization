```python
from pyomo.environ import*
from pyomo.opt import*

model=ConcreteModel()

model.x_OA=Var(within=NonNegativeIntegers)
model.x_OB=Var(within=NonNegativeIntegers)
model.x_OC=Var(within=NonNegativeIntegers)
model.x_AB=Var(within=NonNegativeIntegers)
model.x_AD=Var(within=NonNegativeIntegers)
model.x_BD=Var(within=NonNegativeIntegers)
model.x_BC=Var(within=NonNegativeIntegers)
model.x_BE=Var(within=NonNegativeIntegers)
model.x_CE=Var(within=NonNegativeIntegers)
model.x_ED=Var(within=NonNegativeIntegers)
model.x_ET=Var(within=NonNegativeIntegers)
model.x_DT=Var(within=NonNegativeIntegers)
model.x_TO=Var(within=NonNegativeIntegers)

model.obj= Objective(expr=model.x_TO, sense=maximize)

model.con1= Constraint(expr=-model.x_TO+model.x_OA+model.x_OB+model.x_OC==0)
model.con2= Constraint(expr=-model.x_OA+model.x_AB+model.x_AD==0)
model.con3= Constraint(expr=-model.x_OB-model.x_AB+model.x_BC+model.x_BE+model.x_BD==0)
model.con4= Constraint(expr=-model.x_OC-model.x_BC+model.x_CE==0)
model.con5= Constraint(expr=-model.x_AD-model.x_BD-model.x_ED+model.x_DT==0)
model.con6= Constraint(expr=-model.x_BE-model.x_CE+model.x_ED+model.x_ET==0)
model.con7= Constraint(expr=-model.x_DT-model.x_ET+model.x_TO==0)
model.con8= Constraint(expr=model.x_OA<=5)
model.con9= Constraint(expr=model.x_OB<=7)
model.con10= Constraint(expr=model.x_OC<=4)
model.con11= Constraint(expr=model.x_AB<=1)
model.con12= Constraint(expr=model.x_AD<=3)
model.con13= Constraint(expr=model.x_BD<=4)
model.con14= Constraint(expr=model.x_BC<=2)
model.con15= Constraint(expr=model.x_BE<=5)
model.con16= Constraint(expr=model.x_CE<=4)
model.con17= Constraint(expr=model.x_ED<=1)
model.con18= Constraint(expr=model.x_ET<=6)
model.con19= Constraint(expr=model.x_DT<=9)

Solver=SolverFactory('glpk')
results=Solver.solve(model)
print(results)

print('The value of xOA is', value(model.x_OA))
print('The value of xOB is', value(model.x_OB))
print('The value of xOC is', value(model.x_OC))
print('The value of xAB is', value(model.x_AB))
print('The value of xAD is', value(model.x_AD))
print('The value of xBD is', value(model.x_BD))
print('The value of xBE is', value(model.x_BE))
print('The value of xBC is', value(model.x_BC))
print('The value of xCE is', value(model.x_CE))
print('The value of xDE is', value(model.x_ED))
print('The value of xDT is', value(model.x_DT))
print('The value of xET is', value(model.x_ET))
print('The Maximum is', value(model.obj))

```
